```python
from scraparazzie import scraparazzie
```

```python
client=scraparazzie.NewsClient(language='chinese traditional',location='Taiwan',query='北捷',max_results=10)
client.print_news()
items=client.export_news()
print(len(items))
for i,item in enumerate(items):
  print('第'+str(i+1)+'則新聞:')
  print('新聞標題'+item['title'])
  print('新聞機構'+item['source'])
  print('新聞連結'+item['link'])
  print('新聞時間'+item['publish_date'])
  print('==================================================')
```
